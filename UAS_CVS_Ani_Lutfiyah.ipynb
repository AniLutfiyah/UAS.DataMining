```python
#Nama   = Ani Lutfiyah
#NIM    = 32602000071

!pip install KeyBERT
import re
import pandas as pd
from keybert import KeyBERT

# Load data from Csv
data = pd.read_csv ('/content/abstrak ani.csv')
data = data.fillna('')

# Preprocess text
def preprocess_text(text):
    if isinstance(text, str):
        text = re.sub(r'\d+', '', text)
        text = re.sub(r'[^\w\s]', '', text)
        text = text.lower()
    return text

data['Abstrak'] = data['Abstrak'].apply(preprocess_text)

# Initialize KeyBERT
model = KeyBERT()

# Extract keywords
for index, row in data.iterrows():
    judul = row['Judul']
    abstrak = row['Abstrak']

    keywords = model.extract_keywords(abstrak, keyphrase_ngram_range=(1, 2), stop_words='english', top_n=3)
    keyword_list = [keyword for keyword, score in keywords]

    print('Judul:', judul)
    print('Keywords:', keyword_list)
    print()
```
